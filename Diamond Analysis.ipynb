```python
# Data manipulation and analysis
import numpy as np  # Numerical operations
import pandas as pd  # Data structures and data analysis

import matplotlib.pyplot as plt  # Plotting
import seaborn as sns  # Statistical data visualization


```

```python
   df = pd.read_csv('D:\DA Portfolio Projects\Python Projects\Diamonds\diamonds.csv')
```

```python
df.head()
```

```python
df.rename(columns={
    'x': 'length in mm',
    'y': 'width in mm',
    'z': 'depth in mm'
}, inplace=True)
```

```python
df.head()
```

```python
# Calculate the TOTAL DEPTH PERCENTAGE
df['TOTAL DEPTH PERCENTAGE'] = 2 * df['depth in mm'] / (df['length in mm'] + df['width in mm'])
```

```python
df['TOTAL DEPTH PERCENTAGE']
```

```python
df.head()
```

```python
# Group by 'type' and calculate the mean price for each type
mean_prices = df.groupby('cut')['price'].mean().reset_index()

```

```python
mean_prices.rename(columns={'price': 'average_price'}, inplace=True)
```

```python
most_expensive_types = mean_prices.sort_values(by='average_price', ascending=False)
```

```python
most_expensive_types
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='cut', y='price', data=df)
plt.title('Average Price by Diamond Type')
plt.xlabel('Diamond Type')
plt.ylabel('Average Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='cut', y='price', data=df)
plt.title('Price Distribution by Diamond Type')
plt.xlabel('Diamond Type')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='cut', y='price', data=df)
plt.title('Price Distribution by Diamond Type')
plt.xlabel('Diamond Type')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['carat'], df['depth'], color='blue', alpha=0.5)
plt.title('Carat vs Depth')
plt.xlabel('Carat')
plt.ylabel('Depth')
plt.grid(True)
plt.show()
```
